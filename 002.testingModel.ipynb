```python
import keras
import matplotlib.pyplot as plt
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import os
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from math import sqrt
```

```python
YAHOO_DS="C:\CSU\CS-533\TERM-PROJECT\WS\WORK\Stochastic-Methods\data\yahoo\dataset\ydata-labeled-time-series-anomalies-v1_0"
DIRS = ['A1Benchmark',
 'A2Benchmark',
 'A2Mutation',
 'A3Benchmark',
 'A4Benchmark']
```

```python
#Reading one file and checking the data
df = pd.read_csv(YAHOO_DS + os.path.sep + DIRS[0] + os.path.sep + "real_1.csv") 
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'],unit='s')#format='%f' if formatting required upto nanoseconds
df_indexed = df.set_index('timestamp')
```

```python
values = df_indexed.copy()
values = values.drop(columns=['is_anomaly'],axis=1)
#values = values.reshape((len(values), 1))
# train the standardization
scaler2 = StandardScaler()
scaler2 = scaler2.fit(values)
print('Mean: %f, StandardDeviation: %f' % (scaler2.mean_, sqrt(scaler2.var_)))
# standardize the dataset
standardaized = scaler2.transform(values)

df['std_value'] = standardaized
df_indexed['std_value'] = standardaized
```

```python
values
```

```python
X = values.values.reshape(-1, )
Y = df_indexed['is_anomaly'].values.reshape(-1, )
```

```python
Y = df_indexed['is_anomaly'].values.reshape(-1, )
```

```python
X
```

```python
# Create the model
model = Sequential()
#model.add(Dense(12, input_dim=1, activation='relu'))
model.add(Dense(100, input_dim=1, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model
```

```python
#model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, Y)
print('Accuracy: %.2f' % (accuracy*100))
```
